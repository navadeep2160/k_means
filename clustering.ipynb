```python
#step-1: Importing the Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
#step-2: Load the data set
data=pd.read_csv("/content/drive/MyDrive/Mall_Customers.csv")
```

```python
#Display the columns names
print(data.columns)
print(data.head())
print(data.tail())
```

```python
#step-4: Encoding gender Genere column
le=LabelEncoder()
data['Genre']=le.fit_transform(data['Genre'])
```

```python
#step-5: Features selection
x=data[['Age','Genre','Spending Score (1-100)']]
```

```python
#step-6: Apply K-means clustring
kmeans=KMeans(n_clusters=4,random_state=42)
data['Cluster']=kmeans.fit_predict(x)
```

```python
#step-7: plot
plt.figure(figsize=(8,5))

plt.scatter(
    data['Age'],
    data['Spending Score (1-100)'],
    c=data['Cluster'],
    cmap='viridis'
)

plt.xlabel('Age')
plt.ylabel('Interest (Spending Score)')
plt.title('Customer Segmentation Based on Age, Gender and Interest')
plt.colorbar(label='Cluster')

plt.show()


```
